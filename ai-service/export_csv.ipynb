```python
import os
import pandas as pd
```

```python
base_path = "dataset"
subsets = ['train', 'validation', 'test']
classes = {'RG': 1, 'NRG': 0}
```

```python
data_list = []
counter = 0

for subset in subsets:
    for folder_name, label_int in classes.items():
        # Label name kita bikin kapital (RG / NRG) sesuai CSV lamamu
        label_name = folder_name.upper()
        
        folder_path = os.path.join(base_path, subset, folder_name)
        
        if os.path.exists(folder_path):
            files = sorted(os.listdir(folder_path))
            for f in files:
                # Cek apakah file dari Kaggle (ada prefix kgl_) atau data lama
                source = "Kaggle-Poveda" if f.startswith("kgl_") else "EyePACS-Original"
                ext = f.split('.')[-1]
                
                # Directory path (sesuai format kamu yang pakai / di depan dan belakang)
                dir_path = f"/{base_path}/{subset}/{label_name}/"
                
                data_list.append([
                    counter,     # Index
                    f,           # filename
                    label_name,  # label_name (RG/NRG)
                    label_int,   # label_int (1/0)
                    subset,      # subset (train/val/test)
                    source,      # source
                    ext,         # extension
                    dir_path     # directory_path
                ])
                counter += 1
```

```python
columns = ['index', 'filename', 'label_name', 'label_int', 'subset', 'source', 'extension', 'directory_path']
df = pd.DataFrame(data_list, columns=columns)
```

```python
df.to_csv('metadata_final.csv', index=False)

print(f"Mantap! CSV baru sudah jadi dengan total {len(df)} baris.")
print(df.tail())
```
