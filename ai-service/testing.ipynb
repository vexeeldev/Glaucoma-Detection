```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
import os
```

```python
model = tf.keras.models.load_model('glaucoma_model_v2.keras')
print("✅ Model Berhasil Dimuat!")
folder_path = "picture/noglaucoma/"
file_names = os.listdir(folder_path)

print(f"Ditemukan {len(file_names)} file di folder {folder_path}\n")

for file_name in file_names:
    if file_name.endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, file_name)
 
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        

        prediction = model.predict(img_array, verbose=0) 
        probabilitas = prediction[0][0]

        status = "RG (Glaukoma)" if probabilitas > 0.5 else "NRG (Normal)"
        
        print(f"File: {file_name} | Skor: {probabilitas:.4f} | Hasil: {status}")
```
